```python
from PIL import Image, ImageDraw
import os
import numpy as np
import pandas as pd
```

```python
WINDOW_SIZE = 320
NUM_LEVELS = 10
color = ["red", "black", "yellow", "green", "blue"]
```

```python
def make_image(draw, width, k, l, d, num_of_cell, diff = True):
    coordinate1 = k*width + width/2 - d/2
    coordinate2 = l*width + width/2 - d/2
    random_color = color[np.random.randint(len(color))]
    draw.ellipse((coordinate1,coordinate2, coordinate1 + d, coordinate2 + d), 
                  fill = random_color, outline = random_color)
    fig = ['ellipse', 'rectangle', 'triangle']
    random_fig = 'ellipse'
    for c1 in range(num_of_cell):
        for c2 in range(num_of_cell):
            if c1 != k or c2 != l :
                other_d = np.random.randint(low = d + 10, high = width) 
                if diff:
                    random_fig = fig[np.random.randint(len(fig))] 
                if random_fig == 'ellipse':
                    coordinate1 = c1*width + width/2 - other_d/2
                    coordinate2 = c2*width + width/2 - other_d/2
                    random_color = color[np.random.randint(len(color))]
                    draw.ellipse((coordinate1,coordinate2,coordinate1 + other_d,coordinate2 + other_d), 
                                          fill=random_color, outline=random_color)
                
                if random_fig == 'rectangle':
                    coordinate1 = c1*width + (width - other_d)/2
                    coordinate2 = c2*width + (width - other_d)/2
                    draw.rectangle((coordinate1, coordinate2,coordinate1 + other_d,coordinate2 + other_d),
                                    fill=random_color, outline=random_color)
                
                if random_fig == 'triangle':
                    h = other_d*np.sqrt(3)/2.0 ## высота

                    coordinate11 = c1*width + width/2
                    coordinate12 = c2*width + (width - h)/2

                    coordinate21 = c1*width + (width - other_d)/2
                    coordinate22 = c2*width + width - (width - h)/2

                    coordinate31 = c1*width + (width - other_d)/2 + other_d
                    coordinate32 = c2*width + width - (width - h)/2       

                    draw.polygon((coordinate11, coordinate12,coordinate21,coordinate22,coordinate31,coordinate32),
                                    fill=random_color, outline=random_color)
                
def make_other_image(draw1, draw2, width, d, num_of_cell, diff = True):
    fig = ['ellipse', 'rectangle', 'triangle']
    random_fig1 = 'ellipse'
    random_fig1 = 'ellipse'
    for c1 in range(num_of_cell):
        for c2 in range(num_of_cell):
            if diff:
                random_fig1 = fig[np.random.randint(len(fig))]   
                random_fig2 = fig[np.random.randint(len(fig))]   
                
            # первая картинка            
            random_color = color[np.random.randint(low = 0, high = len(color))]
            other_d = np.random.randint(low = d + 10, high = width-5)
            if random_fig1 == 'ellipse': 
                coordinate1 = c1*width + width/2 - other_d/2
                coordinate2 = c2*width + width/2 - other_d/2
                draw1.ellipse((coordinate1,coordinate2,coordinate1 + other_d,coordinate2 + other_d), 
                                      fill=random_color, outline=random_color)
                
            if random_fig1 == 'rectangle':
                coordinate1 = c1*width + (width - other_d)/2
                coordinate2 = c2*width + (width - other_d)/2
                draw1.rectangle((coordinate1, coordinate2,coordinate1 + other_d,coordinate2 + other_d),
                                fill=random_color, outline=random_color)
                
            if random_fig1 == 'triangle':
                h = other_d*np.sqrt(3)/2.0 ## высота
                
                coordinate11 = c1*width + width/2
                coordinate12 = c2*width + (width - h)/2
                
                coordinate21 = c1*width + (width - other_d)/2
                coordinate22 = c2*width + width - (width - h)/2
                
                coordinate31 = c1*width + (width - other_d)/2 + other_d
                coordinate32 = c2*width + width - (width - h)/2       
                
                draw1.polygon((coordinate11, coordinate12,coordinate21,coordinate22,coordinate31,coordinate32),
                                fill=random_color, outline=random_color)
                
            #вторая картинка            
            random_color = color[np.random.randint(low = 0, high = len(color))]
            other_d = np.random.randint(low = d + 10, high = width-5) 
            if random_fig2 == 'ellipse':                
                coordinate1 = c1*width + width/2 - other_d/2
                coordinate2 = c2*width + width/2 - other_d/2
                draw2.ellipse((coordinate1,coordinate2,coordinate1 + other_d,coordinate2 + other_d), 
                                      fill=random_color, outline=random_color)
            if random_fig2 == 'rectangle':
                coordinate1 = c1*width + (width - other_d)/2
                coordinate2 = c2*width + (width - other_d)/2
                draw2.rectangle((coordinate1, coordinate2,coordinate1 + other_d,coordinate2 + other_d),
                                fill=random_color, outline=random_color)
            if random_fig2 == 'triangle':
                h = other_d*np.sqrt(3)/2.0 ## высота
                
                coordinate11 = c1*width + width/2
                coordinate12 = c2*width + (width - h)/2
                
                coordinate21 = c1*width + (width - other_d)/2
                coordinate22 = c2*width + width - (width - h)/2
                
                coordinate31 = c1*width + (width - other_d)/2 + other_d
                coordinate32 = c2*width + width - (width - h)/2       
                
                draw2.polygon((coordinate11, coordinate12,coordinate21,coordinate22,coordinate31,coordinate32),
                                fill=random_color, outline=random_color)
                
        
                
                
```

```python
def generate(num_of_cell = 3, diff = True, path = "image"):
    small = []   
    

    for i in range(10): 
        width = int(WINDOW_SIZE/num_of_cell)

        image1 = Image.new("RGBA", (WINDOW_SIZE,WINDOW_SIZE), (0,0,0,0))
        image2 = Image.new("RGBA", (WINDOW_SIZE,WINDOW_SIZE), (0,0,0,0))
        image3 = Image.new("RGBA", (WINDOW_SIZE,WINDOW_SIZE), (0,0,0,0))

        draw1 = ImageDraw.Draw(image1)
        draw2 = ImageDraw.Draw(image2)
        draw3 = ImageDraw.Draw(image3)

        num_of_image = np.random.randint(3) 
        d = np.random.randint(low = 5, high = width - 15)   
        k = np.random.randint(num_of_cell)
        l = np.random.randint(num_of_cell)

        if num_of_image == 0:  
            make_image(draw1, width, k, l, d, num_of_cell, diff = True)
            make_other_image(draw2, draw3, width, d,num_of_cell, diff = True)
        elif num_of_image == 1:
            make_image(draw2,width, k, l, d, num_of_cell, diff = True)
            make_other_image(draw1, draw3, width, d, num_of_cell, diff = True)
        else:
            make_image(draw3,width, k, l, d, num_of_cell, diff = True)
            make_other_image(draw1, draw2, width, d, num_of_cell, diff = True)
        small.append(num_of_image)

        try:
            os.mkdir(f"{path}_{num_of_cell}_#_{i}")
        except OSError:
            print("Файл уже существует")

        image1.save(f"{path}_{num_of_cell}_#_{i}" + "/image1.png", "PNG")
        image2.save(f"{path}_{num_of_cell}_#_{i}" + "/image2.png", "PNG")
        image3.save(f"{path}_{num_of_cell}_#_{i}" + "/image3.png", "PNG")    

        del draw1
        del draw2
        del draw3

    return small
```

```python
dataframe = pd.DataFrame(columns=[f"level_{i+1}" for i in range(NUM_LEVELS)])
for i in range(NUM_LEVELS):
    small = generate(num_of_cell = i + 1, diff = True, path = "image")
    dataframe.loc[:, f'level_{i+1}'] = small
    
```
